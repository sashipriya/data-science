```python
import plotly.express as px
import plotly.figure_factory as ff
```

```python
data = data.sort_values(by = 'Date')
data
```

```python
def interactive_plot(df, column_name, title):
  fig = px.line(title = title)
  for i in df['Entity'].unique():
    d = df[df['Entity'] == i]
    fig.add_scatter(x = d['Date'] , y = d[column_name] , name = i )
  fig.show()
```

```python
interactive_plot(data,'Cases','numer of covid cases')
```

```python
interactive_plot(data,'Deaths','Number of Deaths')
```

```python
fig = px.choropleth(data, locations = 'iso_alpha',color = 'Cases',animation_frame = 'month')
fig.show()
```

```python
from fbprophet import Prophet
```

```python
India_df = data[ data['Entity'] == 'India']
India_df
```

```python
India_df = India_df[ [ 'Date','Cases']]
India_df
```

```python
India_df.reset_index(inplace = True,drop = True)
India_df
```

```python
India_df.columns = ['ds','y']
India_df
```

```python
train ,test = India_df[India_df['ds']<= '2020-09-30'],India_df[India_df['ds'] >= '2020-10-01']
```

```python
train
```

```python
test
```

```python
m = Prophet()
m.fit(train)
```

```python
future = m.make_future_dataframe(periods = 30)
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
from fbprophet.plot import plot_plotly,plot_components_plotly
plot_plotly(m,forecast)
```
